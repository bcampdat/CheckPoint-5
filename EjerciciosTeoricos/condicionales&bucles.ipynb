## ***1. ¿Qué es un condicional?***

[Que es un condicional : ▶](https://youtu.be/qCw5kJKBAiM)


    Es aquel permite definir que una parte del código se ejecute sólo si se cumple una condición. Es verdadera
                Si.....( caliento el hielo ) 
                    se derrite 
                y si no
                    (sigue frio)     

En programación se usa 

                if (SI).....(accion calentar el hielo)   
                    consecuencia ( se derrite)
                else ( si no)
                    consecuencia  (sigue frio)

SINTAXIS
   
```
if condicion:
      bloque de código si la condición es verdadera
else:
      bloque de código si la condición es falsa
```


````python
# si el hielo esta caliente
caliento_hielo = True        # VERDADERO
 
if caliento_hielo:
    print('Se derrite')
else:
    print('Sigue frío')

 # SI NO esta caliente
caliento_hielo = False        # FALSO

if caliento_hielo:
    print('Se derrite')
else:
    print('Sigue frío')
````

USOS:



Los usos son muchos porque es la parte principal de nuestro codigo-programa ya que realizara las decisiones que hemos tomado siguiendo las condiciones indicadas y valida nuestros datos tomando medidas si no se cumplen

Aunque parezca simple a medida que toma decisiones segun nuestras condiciones este puede ir resultando mas complejo  e incluso podemos ofrecer mas de una alternativa 

```
if condicion_1:
    # bloque de código si la condición es verdadera
elif condicion_2:
    # bloque de código si la condicion2 es verdadera
else:
    # bloque de código si la condición es falsa
```


````python
role = input("Por favor, elige un rol: ")
# role = 'usuario'
# role = 'admin'
# role = 'otro'

if role == 'admin':  
  auth = 'can access'
elif role == 'usuario': 
  auth = 'Registrate'
else:
  auth = 'cannot access'

print (auth)
````

## ***2. ¿Cuáles son los diferentes tipos de bucles en Python?*** 
## ***¿Por qué son útiles?***        

[Tipos de Bucles :](tipos%20de%20bucles.mp4)

Un bucle o ciclo es aquel que nos permite realizar repeticiones de un fragmento de código hasta que se cumple la condición que lo origino
hay dos tipos :

* while ................. 

* for ... in

while     (mientras....)

    se cumpla la condicion se ejecutara sin fin (ciclo infinito)
    (se usa cuando no se sabe de antemano los ciclos necesarios) 
    por lo que sera indispensable determinar su fin previamente 

for ( por cada .... )        *elemento*          in (en) ....  secuencia

    evaluara la condicion y generara tantos bucles como elementos disponga 

SINTAXIS
La sintaxis básica del ciclo **for** en Python es la siguiente:

```
for elemento in iterable/secuencia :

    <tu bloque de código>   # codigo a ejecutar por cada elemento

```

Siguiendo nuestro ejemplo anterior...

 ⇓  ⇓  ⇓    imaginemos que tenemos una lista de temperaturas de nuestro congelador



````python
temperatura_hielo = [-4, -2, -1, 0, 1] # secuencia de temperaturas

for estado in temperatura_hielo: # revisa cada temperatura de la lista
    if estado > 0: # si la temperatura es mayor que 0
        print('Se derrite')
    else:
        print('Sigue frío') # si la temperatura es menor que 0
````

La sintaxis básica del ciclo **while** en Python es la siguiente:

```
while <condición>:
    < tu bloque de código> 
```

Ejemplo:

 ⇓  ⇓  ⇓    imaginemos que tenemos un juego de adivina el numero secreto....

````python
# el bucle se repite hasta que el usuario acierte el numero secreto

numero_secreto = 5
num = int(input("Elige un número: "))

while num != numero_secreto: # mientras no acierte  el bucle este se repetirá
    print("Has fallado! Inténtalo de nuevo.")
    num = int(input("Elige un número del 0 al 10: "))

print("¡Felicidades! Has adivinado el número.") # si acierta el bucle  fin
print("Gracias por participar")
````

Los bucles son útiles para:
* Automatizar tareas repetitivas y para procesar colecciones de datos.
* Nos facilita la lectura y comprensión del código
* Nos permite controlar el flujo de la ejecucion 
        **break** detiene el bucle inmediatamente
        **continue** se salta el paso del bucle que queramos

Ejemplo: 
    
    imprimir numeros de 1 al 10 y nos saltamos los pares (continue)
